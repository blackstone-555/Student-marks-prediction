```
x = dfnew.groupby('Region')['Profit'].sum()
profitbyregion=[]
for i in x:
  profitbyregion.append(i)
regions=['central','east','south','west']  
from matplotlib import pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(profitbyregion, labels = regions,autopct='%1.2f%%')
plt.show()  
```

```
gk = dfnew.groupby('State')
statesofusa=[]
profitofstates=[]
for State,group in gk:
  statesofusa.append(State)
  profitofstates.append(group['Profit'].sum())
```

```
dfnew['State'].unique()
```

profit in diff states
It will help us to know stores in which states are getting loss

```
statesofusa=[]
profitofstates=[]
for State,group in gk:
  statesofusa.append(State)
  profitofstates.append(group['Profit'].sum())
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.bar(statesofusa,profitofstates)
plt.xticks(rotation=90)
plt.show()
```

Sales in different states


```
salesofstates=[]
for State,group in gk:
  salesofstates.append(group['Sales'].sum())
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.bar(statesofusa,salesofstates)
plt.xticks(rotation=90)
plt.show()
```
